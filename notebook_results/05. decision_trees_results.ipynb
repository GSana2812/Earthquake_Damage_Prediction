```python
import logging
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from src.classifier_model import Classifier
```

```python
train_df = pd.read_csv('datasets/train_data.csv')
test_df = pd.read_csv('datasets/test_data.csv')
```

```python
X_train = train_df.drop(columns=['severe_damage'])
y_train = train_df['severe_damage']

X_test = test_df.drop(columns=['severe_damage'])
y_test = test_df['severe_damage']
```

```python
model = DecisionTreeClassifier(criterion='gini', max_depth=7)
y_pred = Classifier.fit_predict(model, X_train, X_test, y_train)
```

```python
print(Classifier.metrics(y_test, y_pred))
```

```python
Classifier.confusion_matrix(y_test, y_pred)
```
