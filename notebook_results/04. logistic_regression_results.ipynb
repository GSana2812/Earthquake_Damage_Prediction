```python
import logging
import pandas as pd
from src.model import LogisticRegression
from src.classifier_model import Classifier
```

```python
logging.basicConfig(
    filename='logs/logistic_regression.log',
    level=logging.DEBUG,  # Set the logging level (e.g., DEBUG, INFO, WARNING, ERROR, CRITICAL)
    format='%(asctime)s [%(levelname)s] %(message)s',  # Define the log message format
    datefmt='%Y-%m-%d %H:%M:%S'  # Define the date and time format
)
```

```python
train_df = pd.read_csv('datasets/train_data.csv')
test_df = pd.read_csv('datasets/test_data.csv')
```

```python
X_train = train_df.drop(columns=['severe_damage'])
y_train = train_df['severe_damage']

X_test = test_df.drop(columns=['severe_damage'])
y_test = test_df['severe_damage']
```

```python
model = LogisticRegression(lr=0.001, n_iters=50)
y_pred = Classifier.fit_predict(model, X_train, X_test, y_train)
```

```python
# Calculating metrics
print(Classifier.metrics(y_test, y_pred))
```

```python
#Confusion matrix
Classifier.confusion_matrix(y_test, y_pred)
```
